# AfD Kompakt Membership Magazine



This code allows to extract teasers (titles, teaser texts, links) of 'Brennpunkt' articles in the Alternative für Deutschland's (AfD) digital membership magazine 'Kompakt' (https://afdkompakt.de/im-brennpunkt/).

## Preparations

```python
# import relevant packages
import pandas as pd
!pip install scrapethat
from scrapethat import *
```

## Create relevant functions

```python
# create a function to create a dictionary with title, teaser, link
def get_article_info(article):
  try:
      article_data = {
              "title":  article.select_one('a')['title'],
              "teaser": article.select_one('.mh-excerpt').text.strip(),
              "link":   article.select_one('a')['href']
      }
      return(article_data)
  except:
      return{}
```

```python
# create a function which takes an url and processes the url; concatenates new output in data frame
def process_one_page(link):
  t = read_cloud(link)
  teasers = t.select('.mh-posts-list-content')
  return pd.DataFrame(list(map(get_article_info, teasers)))
```

## Retrieve links to different pages of the magazine

```python
# get links with list comprehension
# depending on scope, replace numbers in 'range(1, 5)' with page numbers of interest
page_links = [f"https://afdkompakt.de/im-brennpunkt/page/{x}" for x in range(1, 5)]
```

## Create a data frame with details of all articles on the selected pages

```python
# create df
all_data =  pd.concat(list(map(process_one_page, page_links)), ignore_index=True)
all_data
```
